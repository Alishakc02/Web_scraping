```python

# get URL
page = requests.get("https://en.wikipedia.org/wiki/Main_Page")

# scrape webpage
soup = BeautifulSoup(page.content, 'html.parser')

# display scraped data
print(soup.prettify())
```

```python
list(soup.children)
```

```python
# find all occurrence of p in HTML
# includes HTML tags
print(soup.find_all('p'))
```

```python
# create object
object = soup.find(id="mp-left")

# find tags
items = object.find_all(class_="mp-h2")
result = items[0]

# display tags
print(result.prettify())
```

```python
obj=soup.find(id='mp-dyk')
#find class
item=obj.find_all(class_='dyk-img')
result=item[0]# assign the first element of the list/tuple item to the variable result.

#display
print(result.prettify())
```

```python
obj1=soup.find(id='mp-bottom')
#find class
item1=obj1.find_all(id='mp-other')
result=item1[0]
#display
print(result.prettify())
```

```python
#I want to display only text from mp-other
print(result.get_text())
```

```python
#Tag name
print(result.name)
```

```python
# Returns all attributes as a dictionary
print(result.attrs)
# Get specific attribute value (like id)
print(result['id'])
# Safer way to access attribute
print(result.get('class'))
```

```python
# Find first link inside result
link = result.find('a')

```

```python
# Example: Print all <li> items inside the result
for li in result.find_all('li'):
    print(li.get_text(strip=True))
```

```python
print(result.parent)          # Get the parent tag
print(result.contents)        # Get list of all children (including tags + text)
print(result.children)        # Iterator for direct children only
print(result.descendants)     # Iterator for all nested elements
```
